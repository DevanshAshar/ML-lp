```python
import pandas as pd
import numpy as np
df=pd.read_csv("https://raw.githubusercontent.com/codebasics/py/master/ML/9_decision_tree/salaries.csv")
df.head(10)
```

```python
ip=df.drop("salary_more_then_100k",axis='columns')
tgt=df["salary_more_then_100k"]
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_company=LabelEncoder()
le_job=LabelEncoder()
le_degree=LabelEncoder()
```

```python
ip['company_n']=le_company.fit_transform(ip['company'])
ip['job_n']=le_job.fit_transform(ip['job'])
ip['degree_n']=le_degree.fit_transform(ip['degree'])
ip.head(10)
```

```python
ip_n=ip.drop(['company','job','degree'],axis='columns')
ip_n
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(ip_n,tgt,train_size=0.7)
```

```python
X_test
```

```python
from sklearn import tree
```

```python
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_test,Y_test)
```

```python
model.score(X_test,Y_test)
```
