```python
import pandas as pd
df=pd.read_csv("/content/train.csv")
df.head(10)
```

```python
ip=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Fare','Age'],axis='columns')
ip.head(10)
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le_sex=LabelEncoder()
le_Pclass=LabelEncoder()
le_Embarked=LabelEncoder()
```

```python
ip['Sex_n']=le_sex.fit_transform(ip['Sex'])
ip['Pclass_n']=le_Pclass.fit_transform(ip['Pclass'])
ip['Embarked_n']=le_Embarked.fit_transform(ip['Embarked'])
ip.head(10)
```

```python
ip_n=ip.drop(['Pclass','Sex','Embarked'],axis='columns')
ip_n.head(10)
```

```python
tgt=ip_n['Survived']
tgt
```

```python
ip_n=ip_n.drop(['Survived'],axis='columns')
```

```python
ip_n
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(ip_n,tgt,train_size=0.7)
```

```python
X_train
```

```python
from sklearn import tree
model=tree.DecisionTreeClassifier()
```

```python
model.fit(X_train,Y_train)
```

```python
model.predict(X_test)
```

```python
Y_test
```

```python
model.score(X_test,Y_test)
```
